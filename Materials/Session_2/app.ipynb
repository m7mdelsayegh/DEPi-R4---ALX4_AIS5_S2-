```python
# task3
# calaucaltor program
# ask the user to enter two number
# ask the user to enter the operator
# print the result
# +, -, *, /
# if the operator is not valid print invalid operator


num1 = int(input("Enter the first number: "))
num2 = int(input("Enter the second number: "))
operator = input("Enter the operator: ")


if operator == "+":
    print(num1 + num2)

elif operator == "-":
    print(num1 - num2)

elif operator == "*":
    print(num1 * num2)

elif operator == "/":
    print(num1 / num2)

else:
    print("Invalid operator")
```

```python
# Task 4
# Create a Stop Watch by seconds
import time

seconds = int(input("Enter num of seconds"))

for i in range (seconds, 0, -1):
    second = i %60
    minute = i // 60
    print(f"Seconds left: 00:{minute:02d}:{second:02d}")
    time.sleep(1)

print("Time's up")    
```
